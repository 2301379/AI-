### PromptTemplateの作成例

```python
from langchain_core.prompts import PromptTemplate
# テンプレートのインスタンスを作成
prompt = PromptTemplate.from_template("""
以下の料理のレシピを作成してね。
料理名:{dish}
 """)

# invokeメソッドで変数を埋めてプロンプトを作成
prompt_value = prompt.invoke( {"dish":"カレー"} )
print( prompt_value.to_string ())
```

### ChatPromptTemplateの作成例

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system","ユーザーが入力したレシピを考えてください"),
    ("human","{dish}"), # "human"は"user"ロールに対応
])

prompt_value = prompt.invoke({"dish":"カレー"})
print( prompt_value )
```

## MessagesPlaceholderを使って履歴をプロンプトに含める

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
# 履歴用のリストに書くメッセージのクラスにもimport
from langchain_core.messages import AIMessage, HumanMessage

# MessagesPlaceHolderを使ってプロンプトを作成
prompt = ChatPromptTemplate.from_messages([
    ("system","あなたは親切で優秀なアシスタントです"),
    # "chat_history"というキーワードのプレースホルダー
    MessagesPlaceholder("chat_history",optional=True),
  ("human","{input}")
])
```

### 会話履歴のリストを渡してpromptテンプレートからpromptを作成

```python
prompt_value = prompt.invoke({
 "chat_history": [
          HumanMessage(content="こんちゃーす！どうもレトルトです。ゲーム実況者です"),
          AIMessage(content="こんにちは、レトルトさん。はじめまして")
                 ],
 "input":"私の職業は分かりますか？",
})
```

```python
print(prompt_value)

import pprint
pprint.pprint(prompt_value,indent=2)
```
