## Gemini APIのモデルのimport

```python
import google.generativeai as genai
# colabのユーザーデータ環境にアクセスするモジュールもimport
from google.colab import userdata
# colabに設定したAPIキーをgenaiに設定する
genai.configure(  api_key=userdata.get('GOOGLE_APIKEY'))
```

### APIのモデル名を指定してモデルのインスタンスを生成する


```python
## model = genai.GenerativeModel("gemini-1.5-flash")
model = genai.GenerativeModel("gemini-flash-latest")
```

### モデルにトークン数を算出させる

```python
text = "私はワンピースが大好き。アニメを見てる。"
# モデルのメソッドを使用してAPIを呼び出す
response = model.count_tokens(text)
# 結果を表示
print(f"{text} \n のトークン数は : {response.total_tokens}")
```
